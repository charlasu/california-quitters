```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://www.ratebeer.com/breweries/idaho/12/213/"
```

```python
response = requests.get(url)
page = BeautifulSoup(response.text, "html.parser")
```

```python
containers = page.table.find_all("td")
```

```python
breweries = []
locations = []

for container in containers:
    
    brewery = container.a
    location = container.find("a", class_="filter")
    
    if brewery == None:
        pass
    else:
        breweries.append(brewery.text)
        
    if location == None:
        pass
    else:
        locations.append(location.text)
```

```python
boxes = page.table.find_all("td", style =  ("color:#ccc;text-align:right;"))

dates = []

for box in boxes:
    
    date = box.text
    
    if date == None:
        pass
    else:
        dates.append(date)
```

```python
data = pd.DataFrame({
    
    "Brewery": breweries, 
    "Location": locations, 
    "Est. Year": dates
})

data
```

```python
data.to_csv("../brewery_data/breweries_ID.csv", index=False, header = True)
```
